Variables dependientes e Independientes

```python
dominio = range(100)
```

Relacion

```python
def rel1(x):
    return x**3 - (3*x**2)

def rel2(x):
    return x**3
```

```python
rango1 = []
rango2 = []
for x in dominio:
    rango1.append(rel1(x))
    rango2.append(rel2(x))
```

#### La gráfica más sencilla del mundo

```python
plt.plot(dominio,rango)
```

#### Figura

```python
plt.figure(figsize = (12,8))

plt.plot(dominio,rango1,'g',label='rel1')
plt.plot(dominio,rango2,'r--',label='rel2')

plt.title('Gráfica 1: Relación')

plt.xlabel('Dominio')
plt.ylabel('Rango')

plt.legend()

plt.show()
```

### numpy

```python
import numpy as np
```

```python
#instalar librería numpy
pip install numpy
```

```python
dominio = np.linspace(0,2*np.pi,100)
```

```python
def seno(x):
    return np.sin(x)

def coseno(x):
    return np.cos(x)
```

```python
rango1 = []
rango2 = []
for x in dominio:
    rango1.append(seno(x))
    rango2.append(coseno(x))
```

```python
plt.figure(figsize = (8,8))
plt.title('SENO VS COSENO')
plt.plot(dominio,rango1,'--',label='seno')
plt.plot(dominio,rango2,label='coseno')
plt.legend()
plt.show()
```
